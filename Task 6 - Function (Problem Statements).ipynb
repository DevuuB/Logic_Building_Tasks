Smart Parking Lot Management System

```python
def smart_parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            if parked > 0:
                parked -= 1

        # Track peak usage
        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak)

    if parked > capacity:
        print("Parking Status: Exceeded Capacity ")
    else:
        print("Parking Status: Available ")


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, logs)
```

Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    # Base time calculation (Assume 5 minutes per km)
    eta = distance * 5

    # Traffic adjustment
    if traffic == "Low":
        eta += 5
    elif traffic == "Medium":
        eta += 10
    elif traffic == "High":
        eta += 15

    # Weather adjustment
    if weather == "Rainy":
        eta += 10
    elif weather == "Stormy":
        eta += 20
    elif weather == "Clear":
        eta += 0

    print("Estimated Delivery Time:", eta, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

Movie Theatre Seat Occupancy Analyzer


```python
def seat_occupancy_analyzer(total_seats, booked_seats):
    # Count booked seats
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull ")
        print("Suggestion: Open additional show ")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Monitor bookings")
    else:
        print("Show Status: Seats Available")
        print("Suggestion: No additional show required")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

seat_occupancy_analyzer(total_seats, booked_seats)
```

Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    total = 0

    # Calculate total using loop
    for reading in cpu_readings:
        total += reading

    # Calculate average
    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    # Classify server load
    if average < 50:
        print("Server Status: Normal ")
    elif 50 <= average <= 80:
        print("Server Status: Warning ")
    else:
        print("Server Status: Critical ")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

Smart Classroom Resource Usage Monitor

```python
def classroom_resource_monitor(resource_usage):
    overused = []

    # Assume overuse threshold = 8 hours per day
    threshold = 8

    # Loop through dictionary
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)

    # Display overused resources
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes ")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No ")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(resource_usage)
```

Online Event Registration Capacity Controller


```python
def event_registration_controller(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # Loop through each registration
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed >= capacity:
        print("Registration Status: Closed ")
    else:
        print("Registration Status: Open ")


# Sample Input
capacity = 100
registrations = 105

event_registration_controller(capacity, registrations)
```
