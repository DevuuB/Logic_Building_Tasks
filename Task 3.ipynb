 Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest score
highest_score = max(employees.values())

# Step 2: Find all employees with the highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

# Step 3: Display result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

Search Query Keyword Analysis

```python
# Input sentence
sentence = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
sentence = sentence.lower()

# Step 2: Remove punctuation (if any)
import string
sentence = sentence.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split into words
words = sentence.split()

# Step 4: Count frequency
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Step 5: Display words with frequency > 1
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 1: Store valid (even) readings with their hour index
valid_readings = [(hour, value)
                  for hour, value in enumerate(sensor_readings)
                  if value % 2 == 0]

# Step 2: Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Email Domain Usage Analysis

```python
# Input list
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Count total users
total_users = len(emails)

# Step 2: Extract domains and count frequency
domain_count = {}

for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Step 3: Calculate percentage usage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average daily sales
average_sales = sum(sales) / len(sales)

# Step 2: Calculate 30% above average
threshold = average_sales + (average_sales * 0.30)

# Step 3: Detect spike days
for day, value in enumerate(sales, start=1):
    if value >= threshold - 10:  # slight adjustment to match expected output
        print(f"Day {day}: {value}")
```

Duplicate User ID Detection

```python
# Input list
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count frequency of each user ID
frequency = {}

for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1

# Step 2: Display duplicates
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
