Social Media - Post Engagement Analyzer

```python
def analyze_post(likes):
    total = 0

    for i in likes:
        total += i

    if total >= 1000:
        print("Total Likes:", total)
        print("Post Status: Viral Post")
    else:
        print("Total Likes:", total)
        print("Post Status: Normal Engagement")


# Example
analyze_post([200, 300, 150, 400])
```

Agriculture – Rainfall Adequacy Checker

```python
def check_rainfall(rainfall_list, required_level):
    total = 0

    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total += rain

    # Calculate average
    average = total / len(rainfall_list)

    # Check rainfall status
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)


# Example
check_rainfall([70, 75, 68, 75], 70)
```

Social Media - Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):

    # Convert list to set
    unique_usernames = set(usernames)

    # Check for duplicates
    if len(unique_usernames) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example
detect_duplicate_accounts(["riya123", "aman45", "riya123", "dev99"])
```

E-Learning – Student Attendance Eligibility

```python
def check_attendance(attendance_list):

    total_classes = len(attendance_list)
    attended = 0

    # Loop to count attended classes (1 = Present, 0 = Absent)
    for status in attendance_list:
        if status == 1:
            attended += 1

    # Calculate percentage
    percentage = (attended / total_classes) * 100

    # Check eligibility
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)


# Example (1 = Present, 0 = Absent)
check_attendance([1, 1, 1, 1, 0])
```

Smart Electricity Bill Analyzer

```python
def calculate_bill(units):

    # Calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Classify usage
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example
total_bill, usage_status = calculate_bill(250)

print("Total Bill:", total_bill)
print("Usage Status:", usage_status)
```
