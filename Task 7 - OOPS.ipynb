Smart Light Controller

```python
# Create SmartLight class
class SmartLight:

    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # Default status is OFF

    def turn_on(self):
        self.status = "ON"
        print(f"{self.name} is {self.status}")

    def turn_off(self):
        self.status = "OFF"
        print(f"{self.name} is {self.status}")

    def display_status(self):
        print(f"{self.name} is {self.status}")


# Taking user input
light_name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ").upper()

# Creating object
light = SmartLight(light_name)

# Performing action
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
else:
    print("Invalid Action")
```

Employee ID Card System

```python
# Create Employee class
class Employee:

    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Taking user input
name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
department = input("Enter Department: ")

# Creating object
employee = Employee(name, emp_id, department)

# Display ID Card
employee.display_id_card()
```

Mobile Contact Record

```python
# Problem 3: Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


#  Sample Input
name = input("Contact Name: ")
phone = input("Phone Number: ")

# Creating object
contact1 = Contact(name, phone)

#  Expected Output
contact1.display_contact()
```

Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Sample Input
name = input("Product Name: ")
price = int(input("Price: "))

# Object Creation
product1 = Product(name, price)

# Output
print("\nFormatted Price Tag")
product1.print_price_tag()
```

Movie Rating Display System

```python
# Create Movie class
class Movie:

    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie_details(self):
        print("\nMovie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Taking user input
movie_name = input("Enter Movie Name: ")
rating = float(input("Enter Rating: "))

# Creating object
movie = Movie(movie_name, rating)

# Display Movie Details
movie.display_movie_details()
```

Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Sample Input
name = input("Customer Name: ")
address = input("Address: ")

# Object Creation
delivery1 = Delivery(name, address)

# Output
print()
delivery1.print_details()
```
