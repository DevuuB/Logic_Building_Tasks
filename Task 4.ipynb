Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False

    if amount not in valid_plans:
        print("Invalid plan selected.")
        return False

    return True


while True:
    amount = int(input("Enter recharge amount: "))

    if validate_recharge(amount):
        print("Recharge successful ✅")
        break
    else:
        print("Please try again.\n")
```

Inventory Reorder Alert System

```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


# Example Input
inventory = {
    "Apples": 10,
    "Bananas": 25,
    "Oranges": 5,
    "Mangoes": 18
}

# Function Call
check_inventory(inventory)
```

Student Result Processing System

```python
def process_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


# Example usage
marks = [60, 70, 45, 55, 80]

average, result = process_result(marks)

print("Average Marks:", average)
print("Result:", result)
```

Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra during peak hours

    return fare


# Retry system using while loop
while True:
    try:
        distance = float(input("Enter distance (in km): "))
        peak = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak)
        print(f"Total Fare: ₹{total_fare:.2f}")

    except ValueError:
        print("Please enter valid input.")

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        print("Thank you for using the Cab Fare Calculator!")
        break
```

Employee Attendance Eligibility

```python
def check_eligibility(attendance):
    present_days = 0

    # Count Present days using loop
    for day in attendance:
        if day == "P":
            present_days += 1

    total_days = len(attendance)
    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Example usage
attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]

percentage, result = check_eligibility(attendance_list)

print("Attendance Percentage:", percentage, "%")
print("Status:", result)
```

Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum length should be 8)"

    # Loop through each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    # Check conditions
    if has_digit and has_special:
        return "Strong Password ✅"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"


# Example usage
pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```
